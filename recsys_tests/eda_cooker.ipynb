```python
import numpy as np
import polars as pl
from scipy import sparse as sps
from scipy.sparse import csr_matrix


class BaseItemRecommender:
    """Base class for item-to-item recommenders (ingredient-based)"""

    def __init__(
        self,
        interactions: pl.DataFrame,
        item_col: str = "name",
        feature_col: str = "ingredient",
    ):
        self.item_col = item_col
        self.feature_col = feature_col

        # Create mappings
        self.id2item = np.array(interactions[item_col].unique())
        self.id2feature = np.array(interactions[feature_col].unique())

        self.item2id = {item: idx for idx, item in enumerate(self.id2item)}
        self.feature2id = {feat: idx for idx, feat in enumerate(self.id2feature)}

        # Build item-feature matrix (recipes × ingredients)
        rows = [self.item2id[i] for i in interactions[item_col]]
        cols = [self.feature2id[i] for i in interactions[feature_col]]
        data = [1] * len(rows)

        # Matrix: Items × Features
        self.interactions_matrix = csr_matrix(
            (data, (rows, cols)),
            shape=(len(self.id2item), len(self.id2feature)),
            dtype=np.float32,
        )

        # Transpose: Features × Items
        self.interactions_matrix_t = self.interactions_matrix.T

        # Non-zero counts
        self.row_lengths = np.array(
            self.interactions_matrix.getnnz(axis=1)
        ).ravel()  # ingredients per item
        self.col_lengths = np.array(
            self.interactions_matrix.getnnz(axis=0)
        ).ravel()  # items per ingredient

    def _get_item_features(self, item_idx: int) -> set[int]:
        """Get features (ingredients) for an item"""
        return set(self.id2feature[self.interactions_matrix[item_idx].indices])

    def find_similar_items(self, item_id: str, topn: int = 10) -> list[str]:
        raise NotImplementedError

    def recommend_batch(
        self, item_ids: list[str], topn: int = 10
    ) -> dict[str, list[str]]:
        raise NotImplementedError
```

```python
class EASERecommender(BaseItemRecommender):
    """EASE model for item-to-item recommendations"""

    def __init__(
        self,
        interactions: pl.DataFrame,
        item_col: str = "name",
        feature_col: str = "ingredient",
        reg_weight: float = 500,
    ):
        super().__init__(interactions, item_col, feature_col)
        self.reg_weight = reg_weight
        self.B = None
        self._fit()

    def _fit(self):
        """Fit EASE model"""
        print("Fitting EASE model...")
        X = self.interactions_matrix

        # Gram matrix: X^T @ X (Features × Features)
        G = X.T @ X

        # Add regularization to diagonal
        G += self.reg_weight * sps.identity(G.shape[0], format="csr")

        # Convert to dense for inversion
        G_dense = G.todense()

        # Invert (this is the slow part)
        print("Inverting gram matrix...")
        P = np.linalg.inv(G_dense)

        # Compute B matrix
        B = P / (-np.diag(P))
        np.fill_diagonal(B, 0.0)

        # B is Features × Features
        # To get item similarities, we compute: X @ B @ X^T = Items × Items
        self.B = B
        print("EASE model fitted!")

    def _get_item_scores(self, item_idx: int) -> np.ndarray:
        """Get scores for all items given one query item"""
        # Get item features as sparse vector
        item_vec = self.interactions_matrix[item_idx]  # 1 × Features

        # Compute: item_vec @ B @ X^T = 1 × Items
        scores = (item_vec @ self.B @ self.interactions_matrix.T).toarray().squeeze()

        return scores

    def find_similar_items(self, item_id: str, topn: int = 10) -> list[str]:
        if item_id not in self.item2id:
            return []

        item_idx = self.item2id[item_id]

        # Get similarity scores
        scores = self._get_item_scores(item_idx)

        # Exclude self
        scores[item_idx] = -np.inf

        # Get top-n
        top_indices = np.argpartition(-scores, min(topn, len(scores) - 1))[:topn]
        top_indices = top_indices[np.argsort(-scores[top_indices])]

        return self.id2item[top_indices].tolist()

    def recommend_batch(
        self, item_ids: list[str], topn: int = 10
    ) -> dict[str, list[str]]:
        """Get recommendations for multiple items at once"""
        recommendations = {}

        for item_id in item_ids:
            recommendations[item_id] = self.find_similar_items(item_id, topn)

        return recommendations
```

```python
def train_val_test_split_recipes(
    df_exploded: pl.DataFrame,
    k_core: int = 5,
    val_percentage: float = 0.2,
    test_percentage: float = 0.2,
    seed: int = 42,
    item_col: str = "name",
    feature_col: str = "ingredient",
) -> tuple[pl.DataFrame, pl.DataFrame, pl.DataFrame]:
    """Split into train/val/test in one pass"""

    # K-core filtering ONCE on full dataset
    print(df_exploded)
    recipe_counts = df_exploded.group_by(item_col).agg(
        pl.len().alias("ingredient_count")
    )
    valid_recipes = recipe_counts.filter(pl.col("ingredient_count") >= k_core)[item_col]

    ingredient_counts = df_exploded.group_by(feature_col).agg(
        pl.len().alias("recipe_count")
    )
    valid_ingredients = ingredient_counts.filter(pl.col("recipe_count") >= k_core)[
        feature_col
    ]

    filtered = df_exploded.filter(
        pl.col(item_col).is_in(valid_recipes)
        & pl.col(feature_col).is_in(valid_ingredients)
    )

    # Split into train/val/test in one pass
    result_df = (
        filtered.sample(fraction=1.0, shuffle=True, seed=seed)
        .with_columns(
            random_val=pl.arange(0, pl.len()).over(item_col),
            recipe_count=pl.len().over(item_col),
        )
        .with_columns(norm_random=pl.col("random_val") / pl.col("recipe_count"))
        .with_columns(
            split=pl.when(pl.col("norm_random") < test_percentage)
            .then(pl.lit("test"))
            .when(pl.col("norm_random") < (test_percentage + val_percentage))
            .then(pl.lit("val"))
            .otherwise(pl.lit("train"))
        )
    )

    # Ensure each recipe has at least 1 ingredient in train
    train_counts = (
        result_df.filter(pl.col("split") == "train")
        .group_by(item_col)
        .agg(pl.len().alias("train_count"))
    )
    valid_recipes = train_counts.filter(pl.col("train_count") > 0)[item_col]
    result_df = result_df.filter(pl.col(item_col).is_in(valid_recipes))

    train = result_df.filter(pl.col("split") == "train").select([item_col, feature_col])
    val = result_df.filter(pl.col("split") == "val").select([item_col, feature_col])
    test = result_df.filter(pl.col("split") == "test").select([item_col, feature_col])

    return train, val, test
```

```python
def train_val_test_split_recipes(
    df_exploded: pl.DataFrame,
    k_core: int = 5,
    val_percentage: float = 0.2,
    test_percentage: float = 0.2,
    seed: int = 42,
    item_col: str = "name_id",
    feature_col: str = "ingridient_id",
) -> tuple[pl.DataFrame, pl.DataFrame, pl.DataFrame]:
    """Split into train/val/test in one pass"""

    # K-core filtering ONCE on full dataset
    print(df_exploded)
    df_exploded = df_exploded.with_columns(
        pl.len().over(item_col).alias("ingredient_count")
    )
    # valid_recipes = recipe_counts.filter(pl.col("ingredient_count") >= k_core)[item_col]

    df_exploded = df_exploded.with_columns(
        pl.len().over(feature_col).alias("recipe_count")
    )
    # valid_ingredients = ingredient_counts.filter(pl.col("recipe_count") >= k_core)[feature_col]

    # filtered = df_exploded.filter(
    #     pl.col(item_col).is_in(valid_recipes)
    #     & pl.col(feature_col).is_in(valid_ingredients)
    # )

    # Split into train/val/test in one pass
    result_df = (
        df_exploded.sample(fraction=1.0, shuffle=True, seed=seed)
        .with_columns(
            random_val=pl.arange(0, pl.len()).over(item_col),
            recipe_count=pl.len().over(item_col),
        )
        .with_columns(norm_random=pl.col("random_val") / pl.col("recipe_count"))
        .with_columns(
            split=pl.when(pl.col("norm_random") < test_percentage)
            .then(pl.lit("test"))
            .when(pl.col("norm_random") < (test_percentage + val_percentage))
            .then(pl.lit("val"))
            .otherwise(pl.lit("train"))
        )
    )

    # Ensure each recipe has at least 1 ingredient in train
    train_counts = (
        result_df.filter(pl.col("split") == "train")
        .group_by(item_col)
        .agg(pl.len().alias("train_count"))
    )
    valid_recipes = train_counts.filter(pl.col("train_count") > 0)[item_col]
    result_df = result_df.filter(pl.col(item_col).is_in(valid_recipes))

    train = result_df.filter(pl.col("split") == "train").select([item_col, feature_col])
    val = result_df.filter(pl.col("split") == "val").select([item_col, feature_col])
    test = result_df.filter(pl.col("split") == "test").select([item_col, feature_col])

    return train, val, test
```
