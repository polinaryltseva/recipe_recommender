```python
import pandas as pd
from clean_vv_names import read_names_from_csv, clean_name, write_mapping_to_csv
from typing import List, Tuple
```

```python
names = read_names_from_csv("names_unique_vv.csv")
output_csv = "names_unique_vv_cleaned.csv"

mapping: List[Tuple[str, str]] = []
for name in names:
    cleaned = clean_name(name)
    mapping.append((name, cleaned))

write_mapping_to_csv(mapping, output_csv)
print(f"Записано {len(mapping)} строк в {output_csv}")
```

```python
vv_dataset = pd.read_csv("products_final_categories2.csv")
vv_dataset.head(3)
```

```python
mapping_short = pd.read_csv("names_unique_vv_cleaned.csv", usecols=["original_name", "cleaned_name"])

vv_dataset = vv_dataset.merge(
    mapping_short,
    left_on="name",
    right_on="original_name",
    how="left"
)

vv_dataset["short_name"] = vv_dataset["cleaned_name"].fillna(vv_dataset["name"])
vv_dataset.drop(columns=["original_name", "cleaned_name"], inplace=True)
```
