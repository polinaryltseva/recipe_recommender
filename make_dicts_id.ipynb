```python
pov_to_vv_1_to_1 = pov_to_vv.drop_duplicates("ingredient")
vv_to_pov_1_to_1 = vv_to_pov.drop_duplicates("product_id")
```

Теперь создадим два json, первый из которых будет переводить айдишники ингридиентов Поваренка в айдишнки продуктов Вкусвилла, а второй делает все наоборот

```python
def lookup_pov_id(ingredient_name: str):
    if ingredient_name in name_to_id_exact:
        return name_to_id_exact[ingredient_name]
    return name_to_id_norm.get(norm(ingredient_name))

def to_str_id(x):
    if x is None or (isinstance(x, float) and pd.isna(x)) or (not isinstance(x, float) and pd.isna(x) if hasattr(pd, 'isna') else False):
        return None
    if isinstance(x, numbers.Number):
        return str(int(x)) if x.is_integer() else str(x)
    return str(x)

def norm(s: str) -> str:
    return s.strip().lower()
```

```python
items_path = Path('items_dict (1).json')
```

```python
with open(items_path, 'r', encoding='utf-8') as f:
    id_to_name = json.load(f)

name_to_id_exact = {}
name_to_id_norm = {}

for pov_id, name in id_to_name.items():
    if isinstance(name, str):
        if name not in name_to_id_exact:
            name_to_id_exact[name] = pov_id
        n = norm(name)
        if n not in name_to_id_norm:
            name_to_id_norm[n] = pov_id
```

```python
name_to_id_norm
```

```python
pov2vv = {}
for _, row in pov_to_vv_1_to_1.iterrows():
    ing = str(row['ingredient']).strip()
    pov_id = lookup_pov_id(ing)
    vv_id = to_str_id(row.get('match_id'))
    if pov_id is None or vv_id is None:
        pov2vv[str(pov_id)] = "-1"
    else:
        pov2vv[str(pov_id)] = str(vv_id)

pov2vv
```

```python
out_pov2vv = items_path.with_name('items_dict_pov2vv.json')
out_vv2pov = items_path.with_name('items_dict_vv2pov.json')
```

```python
with open(out_pov2vv, 'w', encoding='utf-8') as f:
    json.dump(pov2vv, f, ensure_ascii=False, indent=2, sort_keys=True)

vv2pov = {}
for _, row in vv_to_pov_1_to_1.iterrows():
    vv_id = to_str_id(row.get('product_id'))
    ing = str(row['match_ingredient']).strip()
    pov_id = lookup_pov_id(ing)
    if vv_id is None or pov_id is None:
        vv2pov[str(vv_id)] = "-1"
    else:
        vv2pov[str(vv_id)] = str(pov_id)

with open(out_vv2pov, 'w', encoding='utf-8') as f:
    json.dump(vv2pov, f, ensure_ascii=False, indent=2, sort_keys=True)

print(f'Записано: {out_pov2vv} ({len(pov2vv)} пар)')
print(f'Записано: {out_vv2pov} ({len(vv2pov)} пар)')
```
